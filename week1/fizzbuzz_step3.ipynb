```python
def generate_fizzbuzz(max_num):
  result_list=[]
  for i in range(1,max_num+1):
    if i%15==0:
      result_list.append("FizzBuzz")
    elif i%3==0:
      result_list.append("Fizz")
    elif i%5==0:
      result_list.append("Buzz")
    else:
      result_list.append(i)
  print(result_list)
```
